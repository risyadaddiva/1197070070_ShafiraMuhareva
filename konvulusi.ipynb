```python
#  untuk mengimport library yg ada di python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
import cv2
```

```python
citra1 = imread(fname="dua.jpg") # membaca citra dari file
print(citra1.shape) # menampilkan dimensi citra

plt.imshow(citra1, cmap='gray') # menampilkan citra
```

```python
# untuk mendefinisikan sebuah kernel dalam bentuk matriks 3x3
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])

# memfilter kernel ke citra 1 dan disimpan ke dalam variabel citraOutput
citraOutput = cv2.filter2D(citra1, -1, kernel)

# menampilkan citra 1 dan 2 menggunakan subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray') # menampilkan gambar input
ax[0].set_title("Citra Input") # menampilkan title
ax[1].imshow(citraOutput, cmap = 'gray') # menampilkan gambar output
ax[1].set_title("Citra Output") # menampilkan title
```
