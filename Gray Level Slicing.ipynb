```python
# untuk meng import libary yang ada pada python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
image_path = "hiji.jpg" # untuk memuat gambar
image = cv2.imread(image_path, 0) # untuk memuat gambar dari image path, dalam sekala keabuan

row, column = image.shape # mendapatkan jumlah baris dan kolom gambar

img1 = np.zeros((row,column),dtype = 'uint8') # membuat array kosong dengan ukuran yg kosong dengan gambar
 

min_range = 10 # rentang nilai minimum
max_range = 60 # rentang nilai maksimum
 

for i in range(row): # melakukan iterasi setiap piksel menggunakan dua loop
    for j in range(column):
        if image[i,j]>min_range and image[i,j]<max_range:
            img1[i,j] = 255 # mengatur piksel menjadi putih (255) jika nilainya berada dalam range
        else:
            img1[i,j] = 0 # mengatur piksel menjadi hitam (0) jika nilainya di luar range
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # membuat sebuah figure dengan ukuran 12x12 dan ukuran axes 2x2
ax = axes.ravel() # untuk mengubah array multidimensi menjadi array satu dimensi

# Menampilkan citra input dan histogram input
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram Input')

# Menampilkan citra output dan histogram output
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
